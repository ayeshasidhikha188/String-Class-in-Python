* Classes provide a means of bundling data and functionality together.

## b) Explain why strings are Immutable by taking a string method as an example.
* strings are <b>immutable</b> , as they stored in different memory locations

```python
a=a.lower()
print(a,id(a)) # id(a)=1585813804624
```

```python
a=a.upper()
print(a,id(a)) # id(a)= 1585813805744
```

* Here 'a', 'a' have different memory locations, hence  strings are  <b>IMMUTABLE</b>
